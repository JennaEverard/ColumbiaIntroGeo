# II. Python Basics

#### **What is Python?**

Python is a powerful and versatile **programming language** that is widely used for tasks ranging from web development to scientific research. It contains a rich library of tools for everything from data analysis to machine learning.

#### **What is a variable?**

A **variable** is like a box where you can store information. You give it a name and then you can use it to store numbers, text, or other types of data. To see an example, run the following code cell:

```python
# creating a numeric variable
number = 7 # this is a variable called 'number' and we store the number 7 in it
print(number) # this will print the value stored in 'number' (7)

# creating a text variable
text = 'Hello!' # this is a variable called 'text' and we store the string 'Hello!' in it
print(text) # this will print the value stored in 'text' (Hello!)
```

#### **Basic data types**

We mentioned that variables can hold different types of data. The types of data we'll be dealing with include:
- **Integers**: whole numbers (i.e. 1, -56, 104)
- **Floats**: numbers with decimal points (i.e. 3.14, 2.5)
- **Strings**: text (i.e. 'Hello!', 'Earth')
- **Lists**: ordered collection of items (i.e. [1, 2, 3], ['apple', 'orange', 5]). Note that a single list can contain multiple data types.

```python
int_var = 5 # creating an integer variable
float_var = 4.928 # creating a float variable
text_var = 'Hello' # creating a text variable
list_var = ['a', 'b', 'c', 1, 2, 3] # creating a list variable

# the following lines of code will print the variables we just created
print(int_var)
print(float_var)
print(text_var)
print(list_var)
```

#### **Basic math operations**

You can do math with numbers in Python, just as you would with a calculator!

```python
# Let's define two numeric variables to start
a = 5
b = 2.5

sum_result = a + b # adds a and b
subtraction_result = a - b # subtracts b from a
product_result = a * b # multiplies a and b
division_result = a / b # divides a by b
remainder_result = a % b # gets the remainder when dividing a by b

# the following lines of code will print those results
print('Sum: ', sum_result) # note how you can add variables into a print statement
print('Difference: ', subtraction_result)
print('Product: ', product_result)
print('Division: ', division_result)
print('Remainder: ', remainder_result)
```

Note that in our math operations above we mixed data types: a was an integer while b was a float. This is allowed in Python, although it's important to realize that Python then converts the integer to a float by default and returns a value that is a float. If we wanted our result as an integer, we would have to do an operation called **casting**:

```python
division_result_int = int(sum_result) # casting to an integer
print(division_result_int) # print result
```

# III. NumPy Basics

#### **What is NumPy?**

NumPy is a Python library that makes working with large datasets and large amounts of numbers easy and efficient. As we saw above, with Python we can store a collection of numbers in a list. With NumPy, we can store numbers in an **array** - a data structure conceptually very similar to a list, but optimized to do math with these numbers incredibly fast.

In order to use such a library, we have to **import** it into our python workspace. When importing libraries, coders often give the library a pseudoname, since you have to re-type the library name each time you call one of its functions. And it's much faster to type 'np' rather than 'NumPy' :)


```python
# Let's import NumPy into our notebooks!
import numpy as np
```

#### **Creating a NumPy array**

An **array** is similar to a list, but it's optimized for performing mathematical operations.

```python
# This is our code to define an array
# The framework is np.array([]) <- the square brackets are important!
array = np.array([1, 2, 3, 4, 5])
print(array)
```

#### **Basic math with arrays**

Sometimes you want to perform the same mathematical operation to every number in your dataset (i.e. maybe you're converting your elevation measurements in meters to kilometers). Rather than manually doing this for each individual number, you can perform math operations on all the elements of an array at once.

First, let's just modify the array by a constant. In this example, we'll multiply each element in our original array by 3.

```python
tripled_array = array * 3 # tripling each element
print(tripled_array) # printing new array
```

Next, let's look at how we can modify the array by another array. A simple example we can do is take the sum of each element in our original array and it's corresponding element in the tripled array (i.e. 0th element + 0th element, 1st element + 1st element, and so on).

```python
summed_array = array + tripled_array # adding the two arrays
print(summed_array) # printing new array
```

This is much faster than doing the same operation with normal Python lists.

Something you may have just noticed is that I referred to the first element in the array as the 0th element. Well **Python is what we call a 0-indexed language**, meaning that when we count in Python, we start from 0. Thus, whenever we're accessing the first value or the first element in a data structure, it's actually at "index 0". The second value is at "index 1", the third at "index 2", and so on.

To access individual elements of a NumPy array, you use square brackets and an index. For example, to get the third element of our original array (which should be 3):

```python
third_elem = array[2] # getting third element - located at index 2
print(third_elem) # printing it
```

# IV. Matplotlib Basics

#### **What is Matplotlib?**

Matplotlib is another powerful Python library, with its specialty being creating visualizations. It is one of the most common ways to plot data on a graph. Let's start by importing it (and installing if necessary) with another short nickname!

```python
import matplotlib.pyplot as plt # IMPORTANT: remember the .pyplot
```

```python
# To install if needed

#pip install numpy
#conda install numpy
```

#### Plotting a Graph

To plot a graph using Matplotlib, we'll follow a few steps:

1. Prepare our x and y data. We'll define two NumPy arrays to do this.
2. **plt.plot()** will allow us to plot the data
3. **plt.xlabel()** and **plt.ylabel()** to label the axes
4. **plt.title()** to give our plot a title
5. **plt.legend()** to add a legend
6. **plt.show()** to show our plot

Let's move into the code:

```python
# step 1: prepare x and y
x = np.array([1,2,3,4,5])
y = x**2 # this squares each element in x

# step 2: plot call.
# note the order of arguments: x, y, style, label
# for style we put 'o-' to specify we want circle markers and lines connecting them
# we have to use the label argument in order to add a legend later
plt.plot(x,y,'o-', label='line plot')

# step 3: label axes
plt.xlabel('x')
plt.ylabel('x-squared')

# step 4: title
plt.title('Our First Plot')

# step 5: legend
plt.legend()

# step 6: display plot
plt.show()
```
