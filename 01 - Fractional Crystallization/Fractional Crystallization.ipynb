```python
# Some helper functions I've provided for you
# Don't worry about understanding all this code! You can just run it and keep working

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def round_to_even(number):
    rounded = round(number)
    if rounded % 2 != 0:
        if number > rounded:
            rounded += 1
        else:
            rounded -= 1
    return rounded

def round_array_to_even(numbers):
    return [round_to_even(num) for num in numbers]

def print_magmas(ofb, con, cab):
    print("OIB:")
    print("SiO2:", ofb[0], " TIO2:", ofb[1], " Al2O3:", ofb[2], " FeO:", ofb[3], " MgO:", ofb[4],
         " CaO:", ofb[5], " Na2O:", ofb[6], " K2O:", ofb[7], " Other:", ofb[8])
    print("Fraction Remaining in Melt: ", np.round(np.sum(ofb) / 600 * 100, 2))
    print("Percent SiO2 (mol): ", np.round(ofb[0] / np.sum(ofb) * 100, 2))
    print()
    print("CON:")
    print(" SiO2:", con[0], " TIO2:", con[1], " Al2O3:", con[2], " FeO:", con[3], " MgO:", con[4],
         " CaO:", con[5], " Na2O:", con[6], " K2O:", con[7], " Other:", con[8])
    print("Fraction Remaining in Melt: ", np.round(np.sum(con) / 600 * 100, 2))
    print("Percent SiO2 (mol): ", np.round(con[0] / np.sum(con) * 100, 2))
    print()
    print("CAB:")
    print(" SiO2:", cab[0], " TIO2:", cab[1], " Al2O3:", cab[2], " FeO:", cab[3], " MgO:", cab[4],
         " CaO:", cab[5], " Na2O:", cab[6], " K2O:", cab[7], " Other:", cab[8])
    print("Fraction Remaining in Melt: ", np.round(np.sum(cab) / 600 * 100, 2))
    print("Percent SiO2 (mol): ", np.round(cab[0] / np.sum(cab) * 100, 2))
    print()

def visualize_magmas(ofb, con, cab, solidify=False):

    total_elems=30*20
    total_ofb_magma = ofb[0] + ofb[1] + ofb[2] + ofb[3] + ofb[4] + ofb[5] + ofb[6] + ofb[7] + ofb[8]
    total_con_magma = con[0] + con[1] + con[2] + con[3] + con[4] + con[5] + con[6] + con[7] + con[8]
    total_cab_magma = cab[0] + cab[1] + cab[2] + cab[3] + cab[4] + cab[5] + cab[6] + cab[7] + cab[8]
    ofb_mat = np.array([1]*ofb[0] + [2]*ofb[1] + [3]*ofb[2] + [4]*ofb[3] + [5]*ofb[4] +
                       [6]*ofb[5] + [7]*ofb[6] + [8]*ofb[7] + [9]*ofb[8])
    np.random.shuffle(ofb_mat)
    con_mat = np.array([1]*con[0] + [2]*con[1] + [3]*con[2] + [4]*con[3] + [5]*con[4] +
                       [6]*con[5] + [7]*con[6] + [8]*con[7] + [9]*con[8])
    np.random.shuffle(con_mat)
    cab_mat = np.array([1]*cab[0] + [2]*cab[1] + [3]*cab[2] + [4]*cab[3] + [5]*cab[4] +
                       [6]*cab[5] + [7]*cab[6] + [8]*cab[7] + [9]*cab[8])
    np.random.shuffle(cab_mat)
    ofb_mat = np.pad(ofb_mat, (0, total_elems - total_ofb_magma), constant_values=9)
    con_mat = np.pad(con_mat, (0, total_elems - total_con_magma), constant_values=9)
    cab_mat = np.pad(cab_mat, (0, total_elems - total_cab_magma), constant_values=9)
    ofb_mat = ofb_mat.reshape(30,20)
    con_mat = con_mat.reshape(30,20)
    cab_mat = cab_mat.reshape(30,20)
    colors=['grey', 'yellow', 'cyan', 'red', 'lime', 'blue', 'orange', 'black', 'white']
    cmap = mcolors.ListedColormap(colors)
    fig, axs = plt.subplots(1,3, figsize=(11,3))
    axs[0].set_title('OFB')
    axs[0].imshow(ofb_mat, cmap=cmap, origin='upper')
    axs[1].set_title('CON')
    axs[1].imshow(con_mat, cmap=cmap, origin='upper')
    axs[2].set_title('CAB')
    cax = axs[2].imshow(cab_mat, cmap=cmap, origin='upper')
    fig,axes = plt.subplots(1,8, figsize=(11,2))
    axes[0].add_patch(plt.Rectangle((0, 0), 1, 1, color="grey"));axes[0].set_xlim(-0.5, 1.5);axes[0].set_ylim(-0.5, 1.5);axes[0].set_title("SiO2");axes[0].axis("off")
    axes[1].add_patch(plt.Rectangle((0, 0), 1, 1, color="yellow"));axes[1].set_xlim(-0.5, 1.5);axes[1].set_ylim(-0.5, 1.5);axes[1].set_title("TiO2");axes[1].axis("off")
    axes[2].add_patch(plt.Rectangle((0, 0), 1, 1, color="cyan"));axes[2].set_xlim(-0.5, 1.5);axes[2].set_ylim(-0.5, 1.5);axes[2].set_title("Al2O3");axes[2].axis("off")
    axes[3].add_patch(plt.Rectangle((0, 0), 1, 1, color="red"));axes[3].set_xlim(-0.5, 1.5);axes[3].set_ylim(-0.5, 1.5);axes[3].set_title("FeO");axes[3].axis("off")
    axes[4].add_patch(plt.Rectangle((0, 0), 1, 1, color="lime"));axes[4].set_xlim(-0.5, 1.5);axes[4].set_ylim(-0.5, 1.5);axes[4].set_title("MgO");axes[4].axis("off")
    axes[5].add_patch(plt.Rectangle((0, 0), 1, 1, color="blue"));axes[5].set_xlim(-0.5, 1.5);axes[5].set_ylim(-0.5, 1.5);axes[5].set_title("CaO");axes[5].axis("off")
    axes[6].add_patch(plt.Rectangle((0, 0), 1, 1, color="orange"));axes[6].set_xlim(-0.5, 1.5);axes[6].set_ylim(-0.5, 1.5);axes[6].set_title("Na2O");axes[6].axis("off")
    axes[7].add_patch(plt.Rectangle((0, 0), 1, 1, color="black"));axes[7].set_xlim(-0.5, 1.5);axes[7].set_ylim(-0.5, 1.5);axes[7].set_title("K2O");axes[7].axis("off")
    plt.tight_layout()
```

For our fractional crystallization investigation, we will look at three magma types:
1. **OFB** - Ocean Floor Basalts. These are large volcanic formations on oceanic plates, often associated with hotspots or mantle plumes.
2. **CON** - Continental Basalts. These are basalts that are associated with continental environments, such as rift zones and large igenous provinces.
3. **CAB** - Calc-Alkaline Basalts. These are often associated with subduction zone settings and are characterized by their geochemical composition including low Fe/Mg ratio due to iron depletion and enrichment in alkali metals (Na and K).

We'll define our starting composition for each of these magmas using results from Julian Pearce's 1976 paper (See the included table in your lab sheet). This paper provides the mean weight percents of major elements for each magma type. Our arrays for each magma type will have weight-percents in the following order:

$[SiO_2, TiO_2, Al_2O_3, FeO, MgO, CaO, Na_2O, K_2O, other]$

The other is just an added value to make sure all the weight percents in a given array sum to 100.
